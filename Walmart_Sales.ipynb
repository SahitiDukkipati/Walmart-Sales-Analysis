```python
#read data from file and handle null values
import pandas as pd
df=pd.read_csv('walmart_sales.csv',na_values='...')
df.head(40)
```

```python
#rename columns with lowercase
df.columns=df.columns.str.lower()
df.columns
df.head(10)
```

```python
#Add new column profit,loss in the stores
df["sales_change"] = df["weekly_sales"].diff()

# Determine if it's profit or loss (based on increase or decrease)
df["status"] = df["sales_change"].apply(lambda x: "Profit" if x > 0 else "Loss")
df
```

```python
df=df.dropna()
df
```

```python
#check the data types and convert the necessary ones
df.dtypes
```

```python
#convert the date (object) to date format
df.loc[:, 'date'] = pd.to_datetime(df['date'], format="%d-%m-%Y")
df
```

```python
df.dtypes
```

```python
#drop columns
df.drop(columns=['temperature','holiday_flag'],inplace=True)
```
